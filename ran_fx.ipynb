Puprose of this notebook is plotting f(x, y) function and show where it defined. Also it is possible to draw dots.

```python
from sympy import *
x, y, z = symbols('x y z')
import math
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

Defining z = f(x, y):

```python
def func(x, y):
    return (x - 1) ** 3 + (y - 2) ** 2 - 3 * x
```

```python
func(x, y)
```

```python
func_with_sqrt = (x - 1) ** 3 + (y - 2) ** 2 - 3 * x + x ** 0.5
```

```python
func_with_sqrt
```

```python
def z(x, y, fx):
    return eval(str(fx))
```

Setting x and y limits:

```python
xx = np.arange(-1, 4, 0.25)
yy = np.arange(-1, 4, 0.25)
```

```python
x_ax, y_ax = np.meshgrid(xx, yy)
```

Defining coordinates of dot in space, for other purposes:

```python
dot_x = 0
dot_y = 2
dot_z = 10
```

Plotting:

```python
fig = plt.figure()
ax_3d = fig.add_subplot(111, projection='3d')
ax_3d.set_xlabel('x')
ax_3d.set_ylabel('y')
ax_3d.set_zlabel('z')
ax_3d.plot_wireframe(x_ax, y_ax, z(x_ax, y_ax, func))


ax_3d.scatter(dot_x, dot_y, dot_z, s=50, color='r', marker='x')
plt.show()
```
